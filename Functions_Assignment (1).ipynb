Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

 Ans:1. map()
   Purpose:Transforms each element in an iterable by applying a given function.
   The result is a new iterable with the transformed elements.

   Syntax:

  map(function, iterable[, iterable2, ...])
   
   Key Features:

   1.Applies the function to each element independently.
   2.Can handle multiple iterables if the function takes multiple arguments.
   3.Returns a map object (an iterator).

 Example:

  # Square each number in a list
  numbers = [1, 2, 3, 4]
  squared = map(lambda x: x ** 2, numbers)
  print(list(squared))  # Output: [1, 4, 9, 16]

  2. filter()
 Purpose:Filters elements in an iterable based on whether they satisfy a condition.
 The result is a subset of the original iterable.

  Syntax:

   filter(function, iterable)

   Key Features:

  1.The function must return a boolean (True or False).
  2.Retains only the elements for which the function returns True.
  3.Returns a filter object (an iterator).

  Example:

   # Filter out even numbers from a list
   numbers = [1, 2, 3, 4, 5, 6]
   evens = filter(lambda x: x % 2 == 0, numbers)
   print(list(evens))  # Output: [2, 4, 6]


    3. reduce()
  Purpose:Aggregates or reduces an iterable into a single cumulative value.
  The result is not a collection but a single value.

   Syntax:

     from functools import reduce
     reduce(function, iterable[, initializer])

  Key Features:

     1.The function must take two arguments and return a single value.
     2.Iteratively applies the function to combine elements.
     3.Requires importing from the functools module.

  Example:

  from functools import reduce

  # Calculate the product of all elements in a list
  numbers = [1, 2, 3, 4]
  product = reduce(lambda x, y: x * y, numbers)
  print(product)  # Output: 24

 Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

   Ans:

```python
def sum_of_evens(numbers):
    """
    Returns the sum of all even numbers in the given list.

    Args:
        numbers (list): A list of integers.

    Returns:
        int: Sum of even numbers.
    """
    # Use a list comprehension to filter and sum even numbers
    return sum(num for num in numbers if num % 2 == 0)


numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_evens(numbers)
print("Sum of even numbers:", result)
```

```python
def reverse_string(s):
    """
    Returns the reverse of the given string.

    Args:
        s (str): The input string.

    Returns:
        str: The reversed string.
    """
    return s[::-1]  # Use slicing to reverse the string

# Example usage
input_string = "Hello, World!"
reversed_string = reverse_string(input_string)
print("Reversed string:", reversed_string)
```


Q4.Write a Python function that checks if a given number is prime or not from 1 to 200.

  Ans:

```python
def is_prime(num):
    """
    Checks if a given number is prime.

    Args:
        num (int): The number to check.

    Returns:
        bool: True if the number is prime, False otherwise.
    """
    if num <= 1:
        return False  # Numbers less than or equal to 1 are not prime

    for i in range(2, int(num ** 0.5) + 1):  # Check divisibility up to sqrt(num)
        if num % i == 0:
            return False  # If divisible, it's not prime
    return True  # If not divisible by any number, it's prime

# Check numbers from 1 to 200
for number in range(1, 201):
    if is_prime(number):
        print(number, "is prime")
```

Q5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

   Ans:

```python
class FibonacciIterator:
    """
    An iterator class that generates the Fibonacci sequence up to a specified number of terms.
    """
    def __init__(self, terms):
        """
        Initialize the iterator with the number of terms.

        Args:
            terms (int): The number of terms in the Fibonacci sequence to generate.
        """
        self.terms = terms  # Store the number of terms
        self.count = 0      # Count how many terms have been generated
        self.a, self.b = 0, 1  # First two Fibonacci numbers

    def __iter__(self):
        # Return the iterator object itself
        return self

    def __next__(self):
        if self.count < self.terms:
            # Generate the next Fibonacci number
            fib = self.a
            self.a, self.b = self.b, self.a + self.b  # Update Fibonacci numbers
            self.count += 1
            return fib
        else:
            raise StopIteration  # Stop the iteration when the desired terms are generated

# Example usage
fibonacci = FibonacciIterator(10)  # Generate the first 10 Fibonacci numbers
for num in fibonacci:
    print(num)
```

Q6.Write a generator function in Python that yields the powers of 2 up to a given exponent.
  
  Ans:

```python
def powers_of_2(exponent):
    """
    A generator function that yields the powers of 2 up to the given exponent.

    Args:
        exponent (int): The exponent up to which powers of 2 should be yielded.

    Yields:
        int: The powers of 2 from 2^0 up to 2^exponent.
    """
    for i in range(exponent + 1):
        yield 2 ** i  # Yield 2 raised to the power of i

# Example usage
for power in powers_of_2(5):
    print(power)
```

Q7. Implement a generator function that reads a file line by line and yields each line as a string.
  
  Ans:

```python
# Generator function to read a file line by line
def read_file_line_by_line(file_path):
    """
    Reads a file line by line using a generator.
    :param file_path: Path to the file to be read.
    :yield: Each line as a string.
    """
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()  # Remove trailing newline characters
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except Exception as e:
        print(f"Error: {e}")

# Step 1: Upload a file to Google Colab
from google.colab import files

uploaded = files.upload()  # This will prompt you to upload a file
file_name = list(uploaded.keys())[0]  # Get the uploaded file name

# Step 2: Call the generator function and process each line
print(f"\nContents of file '{file_name}':\n")

for line in read_file_line_by_line(file_name):
    print(line)

```

Q9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
  Ans:

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 25, 30, 35, 40, 100]

# Use map() to convert Celsius to Fahrenheit
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Display the results
print("Temperatures in Celsius:", celsius_temperatures)
print("Temperatures in Fahrenheit:", fahrenheit_temperatures)
```

Q10.Create a Python program that uses `filter()` to remove all the vowels from a given string.

  Ans:

```python
# Function to check if a character is NOT a vowel
def is_not_vowel(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

# Input string
input_string = "Hello, how are you doing today?"

# Use filter() to remove vowels from the string
result = filter(is_not_vowel, input_string)

# Convert the result back to a string
output_string = ''.join(result)

# Display the result
print("Original String:", input_string)
print("String without vowels:", output_string)
```

Q11.Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
 Order Number  Book title and Auther           Quantity    Price Per Item      
  34587        Learning Python,Mark Lutz          4           40.95
  98762        Programming Python,Mark Lutz       5           56.80
  77226        Head First Python,paul barry       3           32.95
  88112        Einfuhrung in Python3,Bernd Klein  3           24.99

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,-  if the value of the
order is smaller than 100,00 .

Write a Python program using lambda and map.


   Ans:

```python
# Input data as a list of sublists
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfhrung in Python3, Bernd Klein", 3, 24.99]
]

# Lambda function to calculate order total and apply 10 if the total is below 100
calculate_total = lambda order: (order[0], order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0))

# Use map to process the orders
result = list(map(calculate_total, orders))

# Print the result
print(result)
```
